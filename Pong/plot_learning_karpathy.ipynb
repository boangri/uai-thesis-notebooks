```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('learning_history.csv', header=None, names=['episode', 'score', 'avg'])
```

```python
df
```

```python
score = df.score
```

```python
avg = df.avg
```

```python
fig = plt.figure(figsize=(12,8))
ax = plt.axes()
ax.grid(b=True, which='both')
plt.plot(score, label='score')
plt.plot(avg, label='average score')
plt.legend()
plt.ylabel('Разница в счете')
plt.xlabel('Число сыгранных партий(эпизодов)')
plt.title('Кривая обучения агента Pong-v0')
fig.savefig('learning_karpathy.png')
plt.show;
```
